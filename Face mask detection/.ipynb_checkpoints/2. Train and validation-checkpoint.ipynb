```python
import torch
import torchvision
from torchvision import transforms
import numpy as np
import matplotlib.pylab as plt
from torchsummary import summary
import random
from sklearn.metrics import precision_score, recall_score, accuracy_score, f1_score, roc_curve, roc_auc_score, confusion_matrix
import seaborn as sns

random.seed(222)
np.random.seed(222)
torch.cuda.manual_seed(222)
torch.manual_seed(222)

device = torch.device('cuda:0' if torch.cuda.is_available() else 'cpu')
device
```

```python
#Создадим dataloader для train  и validation

train_transform = transforms.Compose([
    
    transforms.RandomHorizontalFlip(p = 0.3),
    
    transforms.RandomApply(torch.nn.ModuleList([
        transforms.RandomRotation(degrees = (-12, 12)),
        transforms.GaussianBlur((3,3)) 
    ]), p = 0.3),
    
    transforms.RandomApply(torch.nn.ModuleList([
    transforms.ColorJitter(),
    ]), p=0.3),
    
    transforms.ToTensor(),
    transforms.Normalize((0.5,0.5,0.5), (0.5,0.5,0.5))
    
])

val_transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize((0.5,0.5,0.5), (0.5,0.5,0.5))
])

train_dataset = torchvision.datasets.ImageFolder('train/', train_transform)
val_dataset = torchvision.datasets.ImageFolder('validation/', val_transform)

batch_size = 10
train_dataloader = torch.utils.data.DataLoader(train_dataset, batch_size = batch_size, shuffle = True)
val_dataloader = torch.utils.data.DataLoader(val_dataset, batch_size = batch_size, shuffle = False)
```

```python
#Создадим модель
model = torchvision.models.resnet50(pretrained=True)
```

```python
model
```

```python
for layer, param in model.named_parameters():
    
    if 'layer4' not in layer:
        param.requires_grad = False

model.fc = torch.nn.Sequential(torch.nn.Linear(2048, 512),
                                 torch.nn.ELU(),
                                 torch.nn.Dropout(0.2),
                                 torch.nn.Linear(512, 2),
                                 torch.nn.Sigmoid())
```

```python
loss = torch.nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(), lr = 0.002, weight_decay = 0.05)
scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(optimizer,
                                                       mode = 'max',
                                                       patience = 5,
                                                       verbose = True,
                                                       factor = 0.5, min_lr=0.00001)
```

```python
model = model.to(device)
```

```python
num_epoch = 30

max_f1 = 0.94
f1_hist_train = []
f1_hist_validation = []

loss_hist_train = []
loss_hist_validation = []


precision_hist_validation = []
recall_hist_validation = []

#Запускаем обучение
for epoch in range(num_epoch):
    
    model.train()
    
    running_loss = 0
    preds_classes = torch.Tensor([]).cuda()
    real_classes = torch.Tensor([]).cuda()
    
    #Проходим по всей обучающей выборке
    for (idx, batch) in enumerate(train_dataloader):
            batch[0] = batch[0].to(device) #Изображения
            batch[1] = batch[1].to(device) #Метки
            
            optimizer.zero_grad()
            
            preds = model.forward(batch[0]) #Прогноз
            loss_value = loss(preds, batch[1]) #Вычисляем loss
            preds_class = preds.argmax(dim = 1)
            
            loss_value.backward() 
            optimizer.step()
            
            
            running_loss += loss_value.item()
            preds_classes = torch.cat((preds_classes, preds_class), dim = 0)
            real_classes = torch.cat((real_classes, batch[1].data), dim = 0)
            
    epoch_loss = running_loss / len(train_dataloader)   
    epoch_f1 = f1_score(preds_classes.cpu(), real_classes.cpu())
    
    f1_hist_train.append(epoch_f1)
    loss_hist_train.append(epoch_loss)
    
    print("{} Loss: {:.4f} f1: {:.4f}".format('train', epoch_loss, epoch_f1))#.item()))
    
    
    #Валидация
    model.eval()
    running_loss = 0
    preds_classes = torch.Tensor([]).cuda()
    real_classes = torch.Tensor([]).cuda()
    
    for (idx, batch) in enumerate(val_dataloader):
            batch[0] = batch[0].to(device)
            batch[1] = batch[1].to(device)
            
            optimizer.zero_grad()
            
            preds = model.forward(batch[0])
            loss_value = loss(preds, batch[1])
            preds_class = preds.argmax(dim = 1)
            
            running_loss += loss_value.item()
            
            preds_classes = torch.cat((preds_classes, preds_class), dim = 0)
            real_classes = torch.cat((real_classes, batch[1].data), dim = 0)

            
    epoch_loss = running_loss / len(val_dataloader)

    epoch_f1 = f1_score(preds_classes.cpu().detach().numpy(), real_classes.cpu().detach().numpy())
    epoch_recall = recall_score(preds_classes.cpu().detach().numpy(), real_classes.cpu().detach().numpy())
    epoch_precision = precision_score(preds_classes.cpu().detach().numpy(), real_classes.cpu().detach().numpy())
    
    f1_hist_validation.append(epoch_f1)
    recall_hist_validation.append(epoch_recall)
    precision_hist_validation.append(epoch_precision)
    scheduler.step(epoch_f1)
    
    if(epoch_f1 >= max_f1):
        torch.save(model.state_dict(), "best_model_f1_"+str(round(epoch_f1, 2)) + '.pth')
        max_f1 = epoch_f1
    
    print("{} Loss: {:.4f} f1: {:.4f}".format('test', epoch_loss, epoch_f1))
    print("Recall: ", round(epoch_recall, 4), ' Precision: ', round(epoch_precision, 4), '\n')
```

```python
plt.figure(figsize = (14, 8))
plt.plot(loss_hist_train, label = 'Train loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

```python
max_f1
```

```python
plt.figure(figsize = (14, 8))
plt.plot(f1_hist_train, label = 'Validation f1')
plt.plot(f1_hist_validation, label = 'Validation f1')
plt.plot(recall_hist_validation, label = 'Validation recall')
plt.plot(precision_hist_validation, label = 'Validation precision')
plt.xlabel('Epoch')
plt.ylabel('F1')
plt.legend()
plt.show()
```

```python
#Test
test_transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize((0.5,0.5,0.5), (0.5,0.5,0.5))
])

test_dataset = torchvision.datasets.ImageFolder('test/', test_transform)


batch_size_test = 4
test_dataloader = torch.utils.data.DataLoader(test_dataset, batch_size = batch_size_test, shuffle = False)
```

```python
model.eval()    

preds_classes = torch.Tensor([]).cuda()
real_classes = torch.Tensor([]).cuda()
for (idx, batch) in enumerate(test_dataloader):
        batch[0] = batch[0].to(device)
        batch[1] = batch[1].to(device)
            
        preds = model.forward(batch[0])
        preds_class = preds.argmax(dim = 1)
        preds_classes = torch.cat((preds_classes, preds_class), dim = 0)
        real_classes = torch.cat((real_classes, batch[1].data), dim = 0)        

f1 = f1_score(real_classes.cpu().detach().numpy(), preds_classes.cpu().detach().numpy())
recall = recall_score(real_classes.cpu().detach().numpy(), preds_classes.cpu().detach().numpy())
precision = precision_score(real_classes.cpu().detach().numpy(), preds_classes.cpu().detach().numpy())
    
print("Result: ")
print("f1: ", round(f1, 2))
print('Recall', round(recall, 2))
print('Precision: ', round(precision, 2))
```

```python
sns.heatmap(confusion_matrix(real_classes.cpu().detach().numpy(), preds_classes.cpu().detach().numpy()), annot=True, fmt='g', cmap=plt.cm.Blues)
plt.xlabel('Predicted label')
plt.ylabel('True label')
plt.show()
```

```python
fpr, tpr, _ = roc_curve(real_classes.cpu().detach().numpy(),  preds_classes.cpu().detach().numpy())
auc = roc_auc_score(real_classes.cpu().detach().numpy(), preds_classes.cpu().detach().numpy())
plt.figure(figsize = (12, 7))
plt.plot(fpr, tpr, label="auc="+str(round(auc, 3)))
plt.legend(loc=4)
plt.show()
```
