```python
#Создадим функцию, которая будет находить в xml файле значения bounding-box-а и возвращать их массивом
def BBOX(file):
    
    x_min = int(file.find('xmin').text)
    x_max = int(file.find('xmax').text)
    y_min = int(file.find('ymin').text)
    y_max = int(file.find('ymax').text)
    
    return [x_min, y_min, x_max, y_max] #формат PyTorch
```

```python
labels_dict = {'with_mask' : 1, 'without_mask' : 0, 'mask_weared_incorrect' : 2}
```

```python
#Также создадим функцию, которая будет искать соответствующую метку класса
def label_of_BBOX(file):
    if(file.find('name').text == 'with_mask'):
        return labels_dict['with_mask']
    elif(file.find('name').text == "mask_weared_incorrect"):
        return labels_dict['mask_weared_incorrect']
    else:
        return labels_dict['without_mask']
```

```python
#Создадим функцию, которая будет собирать целиком метку класса, bounding-box и номер изображения
def generate_info(file):
    with open(file) as f:
        data = f.read() #Считываем файл
        soup = BeautifulSoup(data, 'xml') #Открываем
        objects = soup.find_all('object') #Ищем все объекты в соответствующей информации об изображении
        
        boxes = []
        labels = []
        
        #Теперь проходим по всем найденным объектам
        for i in objects:
            boxes.append(BBOX(i))
            labels.append(label_of_BBOX(i))
        
        #boxes = torch.as_tensor(boxes, dtype = torch.float32)
        #labels = torch.as_tensor(labels, dtype = torch.int64)
        #img_id = torch.tensor([img_id])
        
        #Теперь представим всю полученную информацию в виде словаря
        
        info = {}
        info['boxes'] = boxes
        info['labels'] = labels
        #info['img_id'] = img_id
        
        return info
```

```python
images = sorted(os.listdir("data/images/"))
labels = sorted(os.listdir("data/annotations/"))



assert len(images) == len(labels)
print(len(images), len(labels))
```

```python
#Теперь сформируем наш датасет для классификации

face_images = [] #Полученные изображения
face_labels = [] #соответствующие метки класса

for i, image in enumerate(images):

    image_read = cv2.imread('data/images/' + image, 1) #Читаем изображение
    image_read = cv2.cvtColor(image_read, cv2.COLOR_BGR2RGB) #Переводим в формат rgb
    label_path = os.path.join('data/annotations/', image[0 : -4] + '.xml') #Создаем путь к соответствующему xml файлу
    info = generate_info(label_path) #Получаем всю необходимую информацию об изображении (все лица + метки)
    
    #Проходим по каждому найденному лицу
    for j in range(0, len(info['boxes'])):
        
        face_bbox = info['boxes'][j]
        
        temp_face = image_read[face_bbox[1]:face_bbox[3], face_bbox[0] : face_bbox[2]] #из всего изображения выделяем лицо
        temp_face = cv2.resize(temp_face, (224, 224)) #Устанавливаем размер 224x224
        temp_face = temp_face / 255 #Нормировка
        
        #Добавляем полученное лицо
        face_images.append(temp_face) 
        face_labels.append(info['labels'][j])
```

```python
#Переводим все данные в тензоры
face_images = torch.as_tensor(face_images, dtype = torch.float32)
face_labels = torch.as_tensor(face_labels, dtype = torch.int64)

print(face_images.shape)
print(face_labels.shape)
```

```python
#Посмотрим на распределение классов
temp_labels_dict = {1 : 'with_mask', 0 : 'without_mask', 2 : 'mask_weared_incorrect'}
temp_face_labels = [temp_labels_dict[i] for i in face_labels.numpy()]
sns.countplot(temp_face_labels);
```

```python
#Отрисуем какую-нибудь пару изображение-метка класса
def show_img_labels(data, labels, idx):
    plt.imshow(data[idx])
    plt.title(str(temp_labels_dict[int(labels[idx].numpy())]))
    plt.legend()
    plt.show()

idx = np.random.randint(0, 500)
show_img_labels(face_images, face_labels, idx)
```

```python
#Разделим на train и test
train_img, test_img, train_labels, test_labels = train_test_split(face_images,
                                                                 face_labels,
                                                                 test_size = 0.2,
                                                                 stratify = face_labels)
```

```python
train_img = train_img.permute(0, 3, 1, 2)
test_img = test_img.permute(0, 3, 1, 2)
print(train_img.shape, test_img.shape)
```

```python
#Загрузим обученный ResNet
model = torchvision.models.resnet18(pretrained = True)
for param in model.parameters():
    param.requires_grad = False
    
model.fc = torch.nn.Linear(model.fc.in_features, 3)

device = torch.device('cuda:0' if torch.cuda.is_available() else 'cpu')
model = model.to(device)

loss = torch.nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(), amsgrad = True, lr = 0.0e-3)
```

```python
#Начнем обучение
batch_size = 5

for epoch in range(10):
    
    order = np.random.permutation(len(train_img)-2) #Генерируем случайные индексы
    
    for start_indexes in range(0, len(train_img)-2, batch_size):
        
        optimizer.zero_grad()
        model.train()
        
        batch_indexes = order[start_indexes : start_indexes + batch_size]
        
        X_batch = train_img[batch_indexes]
        y_batch = train_labels[batch_indexes]
        
        X_batch = X_batch.to(device)
        
        print(X_batch.shape)
        preds = model(X_batch)
        
        y_batch = y_batch.to(device)
        
        loss_value = loss(preds, y_batch)
        loss.value.backward()
        
        
        
    if(True):
        
        order_test = np.random.permutation(len(test_img))
        accuracy = 0
        model.eval()
        
        for start_test in range(0, len(test_img), batch_size):

            batch_indexes_test = order_test[start_test : start_test + batch_size]
            
            X_test_batch = test_img[batch_indexes_test].to(device)
            y_test_batch = test_labels[batch_indexes_test].to(device)
            
            test_preds = model.forward(X_test_batch)
            accuracy += (test_preds.argmax(dim = 1) == y_test_batch).float().mean()

        
        accuracy = accuracy / (len(test_img)/batch_size)
        accuracy_test_hist.append(accuracy.data.cpu())

        if((accuracy >= 0.9)):
            torch.save(model.state_dict(), "best_model.pth")
       
        
        print(accuracy)
```
